## Big Data Exercise 3

```python
import pandas as pd
import glob
import os
```

```python
# Task 1


# Gets current working directory
path_to_csv = os.getcwd()
# Finds all files in the directory with a ".csv-ending"
csv_files = glob.glob(os.path.join(path_to_csv, "*.csv"))
# list for storing the dataframes
dataframes = []
# Stores the dataframes in the list
for file in csv_files:
    data = pd.read_csv(file, index_col = 0)
    dataframes.append(data)
# Merges the dataframes together
result_frame = pd.concat(dataframes, axis = 0, ignore_index= True)
# Resets the index
result_frame.reset_index(inplace = True, drop = True)
result_frame
  
```

```python
# Task 2

result_frame.drop(["x_t","perf"], inplace = True, axis =1)
result_frame
```

```python
# Task 3 A

result_frame["Month"] = result_frame["Order Date"].str[:2]
result_frame
```

```python
# Solving task 3 C first. Dropping rows which has nan values in month
result_frame = result_frame.dropna(subset=['Month'])
result_frame.reset_index(inplace = True, drop = True)
```

```python
# Task 3 B
# Converting the values of "Month" to Int64
result_frame["Month"] = pd.to_numeric(result_frame["Month"], errors="coerce").astype("Int64")
```

```python
result_frame
```

```python
# Task 3 D
columns = ['Quantity Ordered', 'Price Each']
result_frame[columns] = result_frame[columns].apply(pd.to_numeric, errors="coerce", axis=1)
result_frame["Sum"] = result_frame["Quantity Ordered"] * result_frame["Price Each"]
```

```python
# Task 3 E
best_months = result_frame.groupby(['Month']).sum()

best_months
```

```python
# Sorting the best months dataframe by Sum, descending
best_months_sorted = best_months.sort_values(by=["Sum"], ascending=False)
```

```python
# Task 3 F
# Saving the dataframes as CSV files
best_months_sorted.to_csv("best_months_sorted.csv")
result_frame.to_csv("result_frame.csv")
```
